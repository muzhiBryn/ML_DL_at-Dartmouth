### Q2 (12 points)

Write a function that takes in one of text corpora and returns a cleaned version of it, where cleaning entails 1) lowercasing 2) removing english stop words (as defined by NLTK), 3) Porter stemmed.

```python
from nltk.corpus import stopwords

porter=nltk.PorterStemmer()
sws = stopwords.words('english')

#WHAT ABOUT ', . /..'
def clean_text(text):
    # write your code here
    cleaned_text = []
    
    for w in text:
        w = w.lower()
        if w not in sws:
            cleaned_text.append(porter.stem(w))
    return cleaned_text
```

### Q8 (20 points  + 10 potential bonus points)
Using the tweet texts from Q7, use pandas and regular expressions to extract hashtags (hashtags starte with #) and user mentions (mentions start with @) that are between 2-5 characters long (not including # or @) and do NOT contain any digits (but can contain any other characters). For bonus points, use regular expressions to extract dates and times being mentioned in the body of tweets. Make your regular expression as general possible, don't just fit it to the data you have. 

Add the columns "hashtags", "mentions" to the dataframe from Q7 and populare them with the hashtags and mentioned extracted from the tweets using your regular expressions. If you are attempting the bonus, include a "date" and "time" columns.  Show your dataframe.

```python
import re
def get_hashtags(t):
    return [''.join(t) for t in re.findall(r"#([^\d\W]{2,5})\W|#([^\d\W]{2,5})$", t)]

def get_mentions(t):
    return [''.join(t) for t in re.findall(r"@([^\d\W]{2,5})\W|@([^\d\W]{2,5})$", t)]

def get_date(t):
    r1 = r'\d{1,2}[/-]\d{2}[/-]\d{2,4}'
    r2 = '\d{1,2}(?:th|TH)?[.,\s]*(?:Jan|Feb|Mar|Apr|May|Jun|Aug|Sep|Oct|Nov|Dec)[a-z]*[.,\s]*\d{0,4}'
    r3 = '(?:Jan|Feb|Mar|Apr|May|Jun|Aug|Sep|Oct|Nov|Dec)[a-z]*[.,\s]*\d{1,2}(?:th|TH)?[.,\s]*\d{0,4}'
    return [''.join(t) for t in re.findall(f"{r1}|{r2}|{r3}", t)]

def get_time(t):
    return re.findall(r'(?:\d{1,2}:){0,2}\d{1,2}\s*(?:am|AM|pm|PM)|(?:\d{1,2}:){1,2}\d{1,2}', t)

df['hashtags'] = df['text'].apply(get_hashtags) 
df['mentions'] = df['text'].apply(get_mentions)
df['date'] = df['text'].apply(get_date)
df['time'] = df['text'].apply(get_time)
df
```
